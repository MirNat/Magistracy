### Выполнила Мирейко Наталья

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
Y = iris.target
target_names = iris.target_names
```

### TSNE

```python
for perplexity in [2, 5, 30, 50, 100]:
    X_tsne = TSNE(n_components=2, perplexity=perplexity, n_iter=5000, learning_rate=50).fit_transform(X)

    for i, target_name in zip([0, 1, 2], target_names):
        plt.scatter(X_tsne[Y == i, 0], X_tsne[Y == i, 1], label=target_name)

    plt.legend(bbox_to_anchor=(1, 1), loc=2)
    plt.title('t-SNE with perplexity %d' % (perplexity))
    plt.show()
```

### PCA

```python
X_pca = PCA(n_components=2).fit_transform(X)

for i, target_name in zip([0, 1, 2], target_names):
    plt.scatter(X_pca[Y == i, 0], X_pca[Y == i, 1], label=target_name)
    
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.title('PCA')
plt.show()
```

Мне больше нравится результат полученный алгоритмом t-sne c perplexity=50, так как объекты из одного класса расположены кучней, сосредоточенней и с сохранением дистанции между объектами.
